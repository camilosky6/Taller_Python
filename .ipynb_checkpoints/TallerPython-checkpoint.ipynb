```python
# pull all artists albums
def getArtista(artist_id):
    r = requests.get(BASE_URL + 'artists/' + artist_id,
    headers=headers,
    params={'include_groups': 'album', 'limit': 50})
    return r.json()

#Obtener todos tracks de un Artista
def getTracks(artist_id):
    r = requests.get(BASE_URL + 'artists/' + artist_id + '/top-tracks?country=SE',
    headers=headers,
    params={'include_groups': 'album', 'limit': 50})
    return r.json()

#Obtener una lista con los json de los artistas y Tracks
def llenarListas(artist_id_list):
    for artista in artist_id_list:
        #Aniade un Artista a la Lista
        aux = getArtista(artista)
        artistas.append(aux)
        #Aniade los Tracks de un Artista a la lista
        aux = getTracks(artista)
        tracks.append(aux)


def setArtista(artista):
    # Se hace la petici√≥n a spotify para traer el artista
    items = []
    items.append(artista['name'])
    items.append(artista['popularity'])
    items.append(artista['type'])
    items.append(artista['uri'])
    items.append(artista['followers']['total'])

    # Se almacenan los resultados en un dataframe utilizando pandas
    data_df = pd.DataFrame.from_dict(items)
   
    # Se agrega la fecha actual al dataframe
    dt = datetime.utcnow()
    dt.replace(tzinfo=timezone.utc).timestamp()
    data_df['fecha_carga'] = dt

    # Se agrega el origen al dataframe
    data_df['origen'] = artista['external_urls']['spotify']

    #Se seleccionan los datos que van para la base de datos
    data_df_db_artists = data_df_db_artists.append(data_df[['name' , 'popularity' , 'type' , 'uri' , 'followers' , 'origen','fecha_carga', 'fecha_carga_segundos']], ignore_index=True)
    print(data_df)

def setTracks(artista):
    listaTracks = artista['tracks'][:3]
    data_df_tracks = pd.DataFrame.from_dict(listaTracks)
    for t in listaTracks:
        # se obtiene el artista del track
        print(t['artists'][0]['name'])
        data_df_tracks['artista'] = t['artists'][0]['name']
        
        # Se obtiene el albun del track
        print(t['album']['name']) 
        data_df_tracks['album_name'] = t['album']['name']

        # Se obtiene la fecha de lanzamiento del track
        print(t['album']['release_date'])
        data_df_tracks['lanzamiento'] = t['album']['release_date']
        data_df_tracks['lanzamiento'] = pd.to_datetime( data_df_tracks['lanzamiento'], format='%Y-%m-%d')

        # Se carga la fecha de carga
        dt = datetime.utcnow()
        dt.replace(tzinfo=timezone.utc).timestamp()
        data_df_tracks['fecha_carga'] = pd.to_datetime(dt)

        # Se carga el origen
        origen_track = data_df_tracks['external_urls'][0]['spotify']
        data_df_tracks['origen'] = origen_track
        print(data_df_tracks['external_urls'][0]['spotify'])
        
def obtenerTodo(artist_id_list):
    for i in artist_id_list:
        artist_json = getArtista(i)
        setArtista(artist_json)
        tracks_json = getTracks(i)
        setTracks(tracks_json)
    
```
